```python
from ultralytics import YOLO

# Gunakan YOLOv8m (Medium) untuk akurasi yang lebih stabil saat orang bergerak
model = YOLO('yolo11m.pt')

# Mulai Training
# data='construction-ppe.yaml' akan otomatis mendownload/mengatur folder
model.train(data='construction-ppe.yaml', epochs=100, imgsz=640, device=0)
```
